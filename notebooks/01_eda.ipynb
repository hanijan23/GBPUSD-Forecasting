```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf

sns.set_style("whitegrid")
```

```python
df = yf.download("GBPUSD=X", start="2000-01-01", end="2026-01-01")
df.head()
```

```python
df.to_csv("../data/gbpusd_raw.csv")
df.shape
```

```python
df = df.dropna()
df = df[['Close']].copy()
df.head()
```

```python
df['log_return'] = np.log(df['Close'] / df['Close'].shift(1))
df.dropna(inplace=True)
df.head()
```

```python
plt.figure(figsize=(12,5))
plt.plot(df['Close'])
plt.title("GBPUSD Exchange Rate (Close Price)")
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(df['log_return'])
plt.title("GBPUSD Log Returns")
plt.show()
```
